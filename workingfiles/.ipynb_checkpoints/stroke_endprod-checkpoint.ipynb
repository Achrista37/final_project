```python
oversample = SMOTE()
eval_df = stroke_data[['gender','age','hypertension','heart_disease','work_type','avg_glucose_level','bmi','stroke']].sample(int(stroke_data.shape[0]*0.2),random_state=42)
train_df = stroke_data.drop(index=eval_df.index)

X_eval_init,y_eval_init = eval_df[['gender','age','hypertension','heart_disease','work_type','avg_glucose_level','bmi']], eval_df['stroke']
X_train_init,y_train_init = train_df[['gender','age','hypertension','heart_disease','work_type','avg_glucose_level','bmi']], train_df['stroke']


X_train_smote, y_train_smote = oversample.fit_resample(X_train_init,y_train_init)


X_eval_smote,y_eval_smote = oversample.fit_resample(X_eval_init,y_eval_init)
```

```python
from sklearn.pipeline import Pipeline
rf_pipeline = Pipeline(steps = [('scale',StandardScaler()),('RF',RandomForestClassifier(random_state=88))])
svm_pipeline = Pipeline(steps = [('scale',StandardScaler()),('SVM',SVC(random_state=88))])
logreg_pipeline = Pipeline(steps = [('scale',StandardScaler()),('LR',LogisticRegression(random_state=88))])
xgb_pipeline = Pipeline(steps = [('scale',StandardScaler()),('XGB',XGBClassifier(random_state=88))])
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_pipeline = Pipeline(steps = [('scale',StandardScaler()),('DT',DecisionTreeClassifier(random_state=88))])
```

```python
rf_cv = cross_val_score(rf_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1')
svm_cv = cross_val_score(svm_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1')
logreg_cv = cross_val_score(logreg_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1')
dt_cv = cross_val_score(dt_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1')
#xgb_cv = cross_val_score(xgb_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1')
```

```python
print('Mean f1 scores:')
print('Random Forest mean f1 :',cross_val_score(rf_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1').mean())
print('SVM mean f1 :',cross_val_score(svm_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1').mean())
print('Logistic Regression mean f1:',cross_val_score(logreg_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1').mean())
#print('XGB mean f1 :',cross_val_score(xgb_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1').mean())
print('Decision Tree mean :',cross_val_score(dt_pipeline,X_train_smote,y_train_smote,cv=10,scoring='f1').mean())
```

```python
rf_pipeline.fit(X_train_smote,y_train_smote)
svm_pipeline.fit(X_train_smote,y_train_smote)
logreg_pipeline.fit(X_train_smote,y_train_smote)
dt_pipeline.fit(X_train_smote,y_train_smote)
xgb_pipeline.fit(X_train_smote,y_train_smote)

rf_pred   =rf_pipeline.predict(X_eval_init)
svm_pred  = svm_pipeline.predict(X_eval_init)
logreg_pred   = logreg_pipeline.predict(X_eval_init)
dt_pred  = dt_pipeline.predict(X_eval_init)
xgb_pred   = xgb_pipeline.predict(X_eval_init)

#rf_cm  = confusion_matrix(y_eval_smote,rf_pred )
#svm_cm = confusion_matrix(y_eval_smote,svm_pred)
#logreg_cm  = confusion_matrix(y_eval_smote,logreg_pred)
#dt_cm = confusion_matrix(y_eval_smote,dt_pred)
#xgb_cm  = confusion_matrix(y_eval_smote,xgb_pred)

rf_f1  = f1_score(y_eval_init,rf_pred)
svm_f1 = f1_score(y_eval_init,svm_pred)
logreg_f1  = f1_score(y_eval_init,logreg_pred)
dt_f1 = f1_score(y_eval_init,dt_pred)
xgb_f1  = f1_score(y_eval_init,xgb_pred)
```
